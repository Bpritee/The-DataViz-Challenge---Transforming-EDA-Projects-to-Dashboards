### Variables Description

* hotel: There are two types of hotel i.e. city or resort hotel

* is_canceled: Booking was canceled or not, where

    1=cancelled

    0=not cancelled.

* lead_time: The number of days between the booking date and the arrival date.

* arrival_date_year: The year of arrival date.

* arrival_date_month: The month of arrival date.

* arrival_date_week_number: The week number of arrival date.

* arrival_date_day_of_month: The day of the month of arrival date.

* stays_in_weekend_nights: The number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel.

* stays_in_week_nights: The number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel.

* adults: The number of adults in the booking.

* children: The number of children in the booking.babies: The number of babies in the booking.

* meal: The type of meal booked, such as Bed and Breakfast, Half board, Full board, or No meal.

* country: Country of the guest.

* market_segment: The market segment designation, such as Direct, Corporate, or Online TA.

* distribution_channel: The distribution channel used to book the reservation.

* is_repeated_guest: Whether the booking guest is a repeated guest, where 1 indicates that the guest has previously stayed at the hotel and 0 indicates that the guest is a new guest.

* previous_cancellations: The number of previous bookings that were canceled by the guest.

* previous_bookings_not_canceled: The number of previous bookings that were not canceled by the guest.

* reserved_room_type: The code of room type reserved by the guest.

* assigned_room_type: The code for the type of room assigned to the booking.
booking_changes: The number of changes made to the booking from the moment the booking was entered into the system until the day of arrival.

* deposit_type: The type of deposit made by the guest, such as No Deposit, Refundable, or Non-Refundable.

* agent: The ID of the travel agency that made the booking.

* company: The ID of the company or entity that made the booking.

* days_in_waiting_list: The number of days the booking was on the waiting list before it was confirmed to the customer.

* customer_type: The type of booking, such as Contract, Group, or Transient.

* adr: The average daily rate, which is the total booking amount divided by the total number of nights stayed.

* required_car_parking_spaces: The number of car parking spaces required by the guest.

* total_of_special_requests: The number of special requests made by the guest.

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Getting the dataset shape
hotel_booking_data.shape
```

```python
# Write your code to make your dataset analysis ready.
# Droping the duplicates value
hotel_booking_data.drop_duplicates(inplace=True)
```

```python
# Checking the Size of data after dropping duplicate
hotel_booking_data.shape
```

```python
# Checking datast data information
hotel_booking_data.info()
```

```python
# Print the blank or null values from dataset
print(hotel_booking_data.isna().sum())
```

```python
# Filling the null or empty values with zero
hotel_booking_data[['country','agent','company','children']]=hotel_booking_data[['country','agent','company','children']].fillna(0)
```

```python
# Checking all null values are eliminated or not
hotel_booking_data.isna().sum()
```

```python
# After filling null values and printing the shape
hotel_booking_data.shape
```

```python
# Changing the datatype to integer
hotel_booking_data[['company', 'agent','children']] = hotel_booking_data[['company', 'agent','children']].astype('int64')
```

```python
# Also changing the datatype of reservation status date also
hotel_booking_data['reservation_status_date']=pd.to_datetime(hotel_booking_data['reservation_status_date'],format='%Y-%m-%d')
```

```python
# Calculating Total Nights Stay in Both the hotels
hotel_booking_data['Total Nights']=hotel_booking_data['stays_in_weekend_nights']+hotel_booking_data['stays_in_week_nights']
```

```python
# After changing checking for shape
hotel_booking_data.shape
```

```python
hotel_booking_data.info()
```

```python
# Calulating total number of peoples stay in hotels
hotel_booking_data['Total_People']=hotel_booking_data['adults']+hotel_booking_data['children']+hotel_booking_data['babies']
```

```python
# Calculating total_revenue
hotel_booking_data['Revenue']=hotel_booking_data['adr']*hotel_booking_data['Total Nights']
```

```python
# Finding the maximum revenue for which type of hotel
max_revenue=hotel_booking_data.groupby('hotel')['Revenue'].max()
max_revenue
```

```python
# Calculating maximum weekend nights in both the hotels
max_weekend_night=hotel_booking_data.groupby('hotel')['stays_in_weekend_nights'].max()
max_weekend_night
```

```python
# Calculating maximum week nights in both the hotels
max_week_night=hotel_booking_data.groupby('hotel')['stays_in_week_nights'].max()
max_week_night
```

```python
# Calculating total nights in both the hotels
max_total_night=hotel_booking_data.groupby('hotel')['Total Nights'].max()
max_total_night
```

```python
# Average daily rates per month for city and resort type hote.
# Grouping hotel and arrival_date_month and finding the average of adr

daily_rates = hotel_booking_data.groupby(["hotel", "arrival_date_month"])["adr"].mean().reset_index(name="avg_daily_rate")

# Sorting the average daily rate in descending order

daily_rates = daily_rates.sort_values(["hotel", "avg_daily_rate"], ascending=False)
print(daily_rates)
```

```python
#Average daily rate by customer type

rate_by_customer_type=hotel_booking_data.groupby(['hotel','customer_type'])['adr'].mean().reset_index(name='avg_daily_rate')
rate_by_customer_type.sort_values(['hotel','avg_daily_rate'],ascending=False)
```

```python
# The total percentage of canceled bookings
canceled_percentage = hotel_booking_data['is_canceled'].mean() * 100
print(canceled_percentage)
```

```python
#From which country the most guest came and number
country_most_guest=hotel_booking_data['country'].value_counts()
country_most_guest
```

```python
# Getting top five from country most guest came
country_most_guest.head()
```
